```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import folium
import branca
```

```python
white_tile = branca.utilities.image_to_url([[1, 1], [1, 1]])
```

```python
latlong = [54.932979810932174, 37.51385698314641]

f = folium.Figure(width=1500, height=750)

m = folium.Map(
    latlong,
    #maxZoom=15,
    #minZoom=5,
    zoom_control=True,
    zoom_start=13,
    scrollWheelZoom=True,
    #maxBounds=[[40, 68],[6, 97]],
    #tiles="Stamen Toner",
    #attr='white tile',
    dragging=True
).add_to(f)
```

```python
descr_lpumg = "Крюковское ЛПУМГ\nКол-во труб категории А3 на балансе завода: 100"

folium.CircleMarker(
    location=latlong,
    radius=15,
    popup=descr_lpumg,
    color="#B83B5E",
    fill=True,
    fill_color="#B83B5E",
).add_to(m)
```

```python
#m.add_child(folium.LatLngPopup())
folium.LayerControl().add_to(m)
m
```
